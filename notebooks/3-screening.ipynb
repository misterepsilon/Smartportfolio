```python
# Load data from SAS pickle file
df = pd.read_parquet(params.PARQUET_PATH, engine="fastparquet")

all = df[(df['DATE'] >= params.START_DATE) & (df['DATE'] <= params.END_DATE)][
    ["DATE", "PERMNO", "PRC", "ME", "RET", "DOLLAR_VOL"]].pivot(index='DATE', columns='PERMNO', values=['PRC', 'ME', "RET", 'DOLLAR_VOL'])
# Extract unique PERMNOs from SAS
permnos = all["PRC"].columns.tolist()
permnos = [permno for permno in permnos if permno not in (np.nan, None) and permno != '']

print(f"Working with {len(permnos)} PERMNOs and a total of {all["PRC"].shape[0]} days over the period {params.START_DATE} to {params.END_DATE}.")
```
